```python
import json
import pandas as pd
import matplotlib.pyplot as plt

txt_file = 'select.txt'
json_file = 'sampla.json'

try:
    # .txtファイルから変数名を読み込む（utf-8-sig を使うとBOM除去が自動）
    with open(txt_file, 'r', encoding='utf-8-sig') as f:
        selected_vars = [line.strip() for line in f if line.strip()]

    print(f"読み込まれた変数: {selected_vars}")
    print(f"変数の数: {len(selected_vars)}")

    # JSONファイルを読み込む
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # DataFrameに変換
    df = pd.DataFrame(data)
    print(f"データベースのカラム: {df.columns.tolist()}")

    # timestampを数値に変換（失敗したら NaN）
    df['timestamp'] = pd.to_numeric(df['timestamp'], errors='coerce')
    df = df.dropna(subset=['timestamp'])

    # 指定された変数 + timestamp を抽出
    cols_to_keep = ['timestamp'] + [col for col in selected_vars if col in df.columns]

    print(f"抽出対象カラム: {cols_to_keep}")
    print(f"見つかったカラム: {[col for col in selected_vars if col in df.columns]}")
    print(f"見つからなかったカラム: {[col for col in selected_vars if col not in df.columns]}")

    df_filtered = df[cols_to_keep].copy()

    # 念のため timestamp 順に並べ替え
    df_filtered = df_filtered.sort_values('timestamp')

    # プロット対象列
    plot_cols = [col for col in df_filtered.columns if col != 'timestamp']
    if not plot_cols:
        print("警告: プロット対象の変数がありません")
        raise SystemExit

    # 値も数値化（欠けている値がある可能性も考慮）
    df_filtered[plot_cols] = df_filtered[plot_cols].apply(pd.to_numeric, errors='coerce')

    # （任意）欠損が「更新されていない＝直前値保持」の意味なら forward fill しておく
    # ※ 欠損が「不明」を意味するならコメントアウトしてください
    df_filtered[plot_cols] = df_filtered[plot_cols].ffill()

    # ====== 要件2：x軸範囲を全プロットで統一 ======
    xmin = df_filtered['timestamp'].min()
    xmax = df_filtered['timestamp'].max()

    x = df_filtered['timestamp'].to_numpy()

    # ====== subplotで縦並び ======
    n = len(plot_cols)
    fig, axes = plt.subplots(
        nrows=n,
        ncols=1,
        sharex=True,
        figsize=(12, 1.2 * n)   # ここは好みで調整
    )

    if n == 1:
        axes = [axes]

    for ax, col in zip(axes, plot_cols):
        y = df_filtered[col].to_numpy()

        # ====== 要件1：直前値保持（ゼロ次ホールド） ======
        ax.step(x, y, where='post')  # 直前の値を次のtimestampまで保持
        # 点も重ねて「観測点」を見やすくする（任意）
        ax.plot(x, y, linestyle='None', marker='o', markersize=2)

        ax.set_title(col)
        ax.set_ylabel("Value")
        ax.grid(True)

        # ====== 要件2：x軸範囲を統一 ======
        ax.set_xlim(xmin, xmax)

    axes[-1].set_xlabel('Timestamp')
    fig.suptitle('Selected Variables over Time (sample-and-hold)')
    fig.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

except FileNotFoundError as e:
    print(f"エラー: {e.filename} が見つかりません。")
except Exception as e:
    print(f"エラーが発生しました: {e}")
```
